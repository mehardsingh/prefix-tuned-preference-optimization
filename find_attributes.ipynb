```python
attribute_names = list(attributes.keys())
att_freq = dict()
for i in range(len(attributes)):
    attribute = attribute_names[i]
    options = set(df[attribute])
    options = {o for o in options if o > 0}
    att_freq[attribute] = len(options)
    print(attribute, options)
    input()
    
with open("att_freq.json", mode="w") as f:
    json.dump(att_freq, f, indent=4)
```
